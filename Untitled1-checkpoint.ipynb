```python
#importing libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
from scipy import stats 

#loading the dataset
df = pd.read_csv(r'C:\Users\pranay\OneDrive\Documents\datasets\multivariate/HBAT.csv')


#a. The histogram of X6, X7, x16 and X17 and the associated normal curve.
x6 = df['x6']
x7 = df['x7']
x16 = df['x16']
x17 = df['x17']

#plotting the histogras
plt.hist(x6, bins=20, color='red', alpha=0.5)
plt.hist(x7, bins=20, color='blue', alpha=0.5)
plt.hist(x16, bins=20, color='green', alpha=0.5)
plt.hist(x17, bins=20, color='yellow', alpha=0.5)

#plotting the normal curves
mu6, sigma6 = stats.norm.fit(x6)
mu7, sigma7 = stats.norm.fit(x7)
mu16, sigma16 = stats.norm.fit(x16)
mu17, sigma17 = stats.norm.fit(x17)

xmin6, xmax6 = plt.xlim()
xmin7, xmax7 = plt.xlim()
xmin16, xmax16 = plt.xlim()
xmin17, xmax17 = plt.xlim()

x6 = np.linspace(xmin6, xmax6, 100)
x7 = np.linspace(xmin7, xmax7, 100)
x16 = np.linspace(xmin16, xmax16, 100)
x17 = np.linspace(xmin17, xmax17, 100)

p6 = stats.norm.pdf(x6, mu6, sigma6)
p7 = stats.norm.pdf(x7, mu7, sigma7)
p16 = stats.norm.pdf(x16, mu16, sigma16)
p17 = stats.norm.pdf(x17, mu17, sigma17)

plt.plot(x6, p6, 'r', linewidth=2)
plt.plot(x7, p7, 'b', linewidth=2)
plt.plot(x16, p16, 'g', linewidth=2)
plt.plot(x17, p17, 'y', linewidth=2)
plt.savefig('plot1.png')

plt.show()

#b. For part a, which variables satisfy the normality test. Examine this visually and using skewness 
#and kurtosis tests.
#calculating skewness and kurtosis
skew6 = stats.skew(x6)
kurt6 = stats.kurtosis(x6)
skew7 = stats.skew(x7)
kurt7 = stats.kurtosis(x7)
skew16 = stats.skew(x16)
kurt16 = stats.kurtosis(x16)
skew17 = stats.skew(x17)
kurt17 = stats.kurtosis(x17)

#printing the results
print("Skewness of X6:", skew6)
print("Kurtosis of X6:", kurt6)
print("Skewness of X7:", skew7)
print("Kurtosis of X7:", kurt7)
print("Skewness of X16:", skew16)
print("Kurtosis of X16:", kurt16)
print("Skewness of X17:", skew17)
print("Kurtosis of X17:", kurt17)

#c. The Normal Probability Plots for X6, X7, x16 and X17. Which variables are normal?
#plotting the normal probability plots
stats.probplot(x6, dist="norm", plot=plt)
plt.title("Normal Probability Plot for X6")
plt.show()
plt.savefig('plot2.png')


stats.probplot(x7, dist="norm", plot=plt)
plt.title("Normal Probability Plot for X7")
plt.show()
plt.savefig('plot3.png')


stats.probplot(x16, dist="norm", plot=plt)
plt.title("Normal Probability Plot for X16")
plt.show()
plt.savefig('plot4.png')

stats.probplot(x17, dist="norm", plot=plt)
plt.title("Normal Probability Plot for X17")
plt.show()
plt.savefig('plot5.png')


#d. Plot the scatterplots for (X7 versus X19) and (X6 versus X19). Are there any outliers and how 
#many? Use an ellipse representing the 95% confidence interval.
x7_19 = df[['x7','x19']]
x6_19 = df[['x6','x19']]

#plotting the scatterplots
plt.scatter(x7_19['x7'], x7_19['x19'])
plt.title("Scatterplot of X7 vs X19")
plt.xlabel("X7")
plt.ylabel("X19")
plt.show()
plt.savefig('plot6.png')


plt.scatter(x6_19['x6'], x6_19['x19'])
plt.title("Scatterplot of X6 vs X19")
plt.xlabel("X6")
plt.ylabel("X19")
plt.show()
plt.savefig('plot7.png')


#calculating the 95% confidence interval
x7_19_mean = x7_19.mean()
x7_19_std = x7_19.std()
x6_19_mean = x6_19.mean()
x6_19_std = x6_19.std()

x7_19_ci = [x7_19_mean - 1.96*x7_19_std, x7_19_mean + 1.96*x7_19_std]
x6_19_ci = [x6_19_mean - 1.96*x6_19_std, x6_19_mean + 1.96*x6_19_std]

#plotting the ellipses
ellipse7 = patches.Ellipse(xy=(x7_19_mean[0], x7_19_mean[1]), width=x7_19_ci[0][0], height=x7_19_ci[1][1], angle=0, edgecolor='r', fc='None', lw=2)
ax = plt.gca()
ax.add_patch(ellipse7)
plt.scatter(x7_19['x7'], x7_19['x19'])
plt.title("Scatterplot of X7 vs X19 with 95% CI Ellipse")
plt.xlabel("X7")
plt.ylabel("X19")
plt.show()
plt.savefig('plot8.png')


ellipse6 = patches.Ellipse(xy=(x6_19_mean[0], x6_19_mean[1]), width=x6_19_ci[0][0], height=x6_19_ci[1][1], angle=0, edgecolor='r', fc='None', lw=2)
ax = plt.gca()
ax.add_patch(ellipse6)
plt.scatter(x6_19['x6'], x6_19['x19'])
plt.title("Scatterplot of X6 vs X19 with 95% CI Ellipse")
plt.xlabel("x6")
plt.ylabel("x19")
plt.show()
plt.savefig('plot9.png')


#e. The bivariate profiling (correlation) of relationships between X6, X7, X8, X12. Which variables 
#are correlated?
#calculating the correlation matrix
corr_matrix = df[['x6','x7','x8','x12']].corr()
print(corr_matrix)
```

```python
corr_threshold = 0.5

# extract upper triangular part of the matrix
upper_triangular = np.triu(corr_matrix, k=1)

# loop through cells and check their values
for i in range(len(upper_triangular)):
    for j in range(i+1, len(upper_triangular)):
        if abs(upper_triangular[i][j]) >= corr_threshold:
            print(f'{corr_matrix.columns[i]} and {corr_matrix.columns[j]} are correlated: {upper_triangular[i][j]}')






```
