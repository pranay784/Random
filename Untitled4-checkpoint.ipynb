```python
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np


# Define the response variable and the predictor variables
df = pd.read_csv(r'C:\Users\pranay\OneDrive\Documents\datasets\multivariate/HBAT.csv')

y = df['x4']
X = df[['x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15']]

# Scale the predictor variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform forward selection
selected_vars = []
auc_scores = []

for i in range(X_scaled.shape[1]):
    best_auc = 0
    best_var = ''
    for var in range(X_scaled.shape[1]):
        if var not in selected_vars:
            model = LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000)
            model.fit(X_scaled[:,selected_vars + [var]], y)
            y_prob = model.predict_proba(X_scaled[:,selected_vars + [var]])[:,1]
            auc = roc_auc_score(y, y_prob)
            if auc > best_auc:
                best_auc = auc
                best_var = var
    selected_vars.append(best_var)
    auc_scores.append(best_auc)

    print("Step {}: Variable {} added to model. AUC = {:.4f}".format(i+1, best_var, best_auc))

# Fit the final model with all selected variables
model_final = LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000)
model_final.fit(X_scaled[:,selected_vars], y)

# Print the summary of the final model
print("\nFinal Model Summary:")
print("Selected variables:", selected_vars)
print("Coefficients:", model_final.coef_)
print("Intercept:", model_final.intercept_)

# Calculate and plot the ROC curve for the final model
y_prob_final = model_final.predict_proba(X_scaled[:,selected_vars])[:,1]
fpr, tpr, _ = roc_curve(y, y_prob_final)
auc_final = roc_auc_score(y, y_prob_final)
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (AUC = {:.4f})'.format(auc_final))
plt.show()








# Perform backward selection
while True:
    model = LogisticRegression(penalty='l2', solver='liblinear')
    model.fit(X, y)
    p_values = pd.Series(model.coef_[0], index=X.columns).apply(lambda x: "{:.5f}".format(x))
    max_p_value = p_values.astype(float).idxmax()
    if float(p_values[max_p_value]) > 0.05:
        X = X.drop(columns=max_p_value)
        print("Variable {} eliminated. Remaining variables: {}".format(max_p_value, list(X.columns)))
        model.fit(X, y)  # Fit model with updated X
        y_prob = model.predict_proba(X)[:,1]
        auc = roc_auc_score(y, y_prob)
        print("AUC = {:.4f}".format(auc))
    else:
        break

# Fit the final model with all selected variables
model_final = LogisticRegression(penalty='l2', solver='liblinear')
model_final.fit(X, y)

# Print the summary of the final model
print("\nFinal Model Summary:")
print("Selected variables:", list(X.columns))
print("Coefficients:", model_final.coef_)
print("Intercept:", model_final.intercept_)

# Calculate and plot the ROC curve for the final model
y_prob_final1 = model_final.predict_proba(X)[:,1]
fpr1, tpr1, _ = roc_curve(y, y_prob_final1)
auc_final1 = roc_auc_score(y, y_prob_final1)
plt.plot(fpr1, tpr1)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (AUC = {:.4f})'.format(auc_final1))



plt.show()


```
